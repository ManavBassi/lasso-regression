```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression 
from sklearn.linear_model import Lasso
from sklearn.datasets import make_regression 
from sklearn.model_selection import train_test_split
```

```python
x,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise =20 ,random_state=13)
x.shape
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
plt.scatter(x,y)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
print(reg.coef_)
print(reg.intercept_)
```

```python
alphas=[0.0001,1,5,10,30]
plt.figure(figsize=(12,6))
plt.scatter(x,y)
for i in alphas:
    l=Lasso(alpha=i)
    l.fit(x_train,y_train)
    plt.plot(x_test,l.predict(x_test),label='alpha={}'.format(i))
plt.legend()
plt.show()
```

```python
m = 100
x1 = 5 *  np.random.rand(m, 1) - 2
x2 = 0.7 * x1 ** 2 - 2 * x1 + 3 + np.random.randn(m, 1)

plt.scatter(x1, x2)
plt.show()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
def gets_preds_lasso(x1,x2,alpha):
    model = make_pipeline(StandardScaler(),PolynomialFeatures(degree=10),Lasso(alpha=alpha,max_iter=1000))
    model.fit(x1,x2)
    return model.predict(x1)

alphas = [0.0001,0.1,1]
cs = ['r','g','b']
plt.figure(figsize=(10, 6))
plt.plot(x1, x2, 'b+', label='Datapoints')

for alpha,c in zip(alphas,cs):
    preds=gets_preds_lasso(x1,x2,alpha)
    plt.plot(sorted(x1[:,0]),preds[np.argsort(x1[:,0])],c,label='alphas={}'.format(alpha))
plt.legend()
plt.show()
```
